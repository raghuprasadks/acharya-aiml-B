Write a function to check a given number is odd or even
pass number as an argument to the function


```python
def evenOrOdd(num):
  if (num%2==0):
    print(num," is even")
  else:
    print(num," is odd")
```

Write shopping function to calculate
discount and amount to pay as per the given details
purchase_amt<=10000 rate=10%
purchase_amt>10000 and purchase_amt<=20000 rate=20%
purchase_amt>20000 rate = 25%
def shopping(purchase_amt)

```python
def shopping(purchase_amt):
  if (purchase_amt<=10000):
    rate=.1
  elif(purchase_amt>10000 and purchase_amt<=20000):
    rate=.2
  else:
    rate=.25
  discount = purchase_amt*rate
  net_amt = purchase_amt-discount
  print(" net amount to pay ",net_amt)
  print(" discount is ",discount)
```

```python
shopping(10000)
```

```python
"""
class and object
"""
class Product():
  def __init__(self,code,name,supplier,price,rating=None):
    self.code = code
    self.name = name
    self.supplier = supplier
    self.price = price
  def information(self):
    print("code : ",self.code," name : ",self.name, " price ",self.price)


```

```python
laptop = Product(1,"HP gen 11","HP",75000)
print(laptop.information())
mobile = Product(2,"Nokia Smart","Nokia",35000)
print(mobile.information())
```
