```python
import pandas as pd
import matplotlib.pyplot as plt
print("data frame")
topeconomy = {
              "country":["USA","CHINA","GERMANY","JAPAN","INDIA"],
              "GDP":[27,18,4.7,4.2,4.1],
              "RANK":[1,2,3,4,5]
              }
df = pd.DataFrame(topeconomy)
```

```python
print("line chart")
plt.title("top economies")
plt.xlabel("Country")
plt.ylabel("GDP")
plt.plot(topeconomy["country"],topeconomy["GDP"])
```

```python
print("scatter chart")
plt.title("top economies")
plt.xlabel("Country")
plt.ylabel("GDP")
plt.scatter(topeconomy["country"],topeconomy["GDP"])
```

```python
print("bar chart")
plt.title("top economies")
plt.xlabel("Country")
plt.ylabel("GDP")
plt.bar(topeconomy["country"],topeconomy["GDP"])
```

```python
print("pie chart")
plt.title("top economies")
plt.pie(topeconomy["GDP"],labels=topeconomy["country"])
```
